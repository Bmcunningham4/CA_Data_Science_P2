```python
import pandas as pd
salary_df = pd.read_csv("Salary.csv") #! Maybe instead of importing these from csv every time import from a diff file, but I guess that's the same thing..
male_df = salary_df[salary_df.Gender == "Male"]
female_df = salary_df[salary_df.Gender == "Female"]
```

```python
diff_jobs_count = salary_df["Job Title"].nunique()
print(diff_jobs_count)  #? There's 129 different jobs so I can't do mean for all of them have to somehow make 5 or so different categories..

diff_jobs = salary_df["Job Title"].unique()
for job in diff_jobs:
    print(job)
```
